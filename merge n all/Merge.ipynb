```python
import pandas as pd 
import numpy as np 
from pandas import Series,DataFrame
```

```python
df1 = DataFrame({'data_set1':np.arange(6),'key':list('aabbcc')})  
df1
```

```python
df2 = DataFrame({'data_set2': np.arange(3) ,'key':list('xbc')})  
df2
```

# Merge (Many to one )

```python
# it automatically choose the merging columns 
# By default (inner join)

pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key') # will get same result || inner pick out only common keys
```

```python
pd.merge(df1,df2,on='key',how='outer') 
```

# Many to Many 

many instance of key in both  DataFrame

It given out cartesian product of the two dataframes 

```python
df3 = DataFrame({'key': list('bbacab'),
               'data_1':np.arange(6)})
df3
```

```python
df4 = DataFrame({'key':list('ababd'),
               'data_2' : np.arange(5)})
df4
```

```python
pd.merge(df3,df4,on='key')  # cartesian product 
```

```python
pd.merge(df3,df4,on='key',how='outer')
```

Merge on multiple keys

```python
dfleft = DataFrame({'key1':['foo','foo','bar'],
                    'key2':['one','two','one'],
                    'lval' : [1,2,3]})
dfleft
```

```python
dfright = DataFrame({'key1':['foo','foo','bar','bar'],
                     'key2':['one','one','two','two'],
                     'rval':[4,5,6,7]})
dfright
```

```python
pd.merge(dfleft,dfright,on=['key1','key2'])  # inner
```

```python
pd.merge(dfleft,dfright,on=['key1','key2'],how='outer')
```

```python
pd.merge(dfleft,dfright,on='key1')  # it will automatically assign the suffix to the keys 
                                    # consider only foo as the key 
```

```python
pd.merge(dfleft,dfright,on='key1',sort='True') # keys are sorted 
```
