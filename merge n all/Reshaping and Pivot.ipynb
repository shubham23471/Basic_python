```python
import pandas as pd 
import numpy as np 
from pandas import Series,DataFrame
```

```python
df1 = DataFrame(20 + np.arange(8).reshape(2,4),
              index=pd.Index(['ohio','colorado'],name='state'),
        columns=pd.Index(['one','two','three','four'],name='Alpha'))
df1
```

```python
result = df1.stack()   # columns to rows 
result
```

```python
result.unstack()  # rows to column 
```

```python
result.unstack(0)
```

```python
result.unstack('state')  # same as result.unstack(0)
```

```python
dfr = DataFrame({'left': result,'right': result + 5},
               
            columns=pd.Index(['left','right'],name="column names"))
dfr
```

```python
dfr.unstack()  # by deafault this will consider the inner most 
```

```python
dfr.unstack('state')
```

```python
dfr.unstack('state').stack('column names')
```
