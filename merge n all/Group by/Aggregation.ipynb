```python
# own data aggregation function 

def max_to_min(arr):
    return max(arr)-min(arr)
```

```python
wino = df_wine.groupby('quality')
```

```python
wino.describe()
```

```python
# unable to view properly in above . so now let's stack it 
wino.describe().stack()
```

```python
# to apply our own funtion we have to pass it as an argument to pandas agg() funtion 

wino.agg(max_to_min)
```

```python
wino.agg('mean') # to get mean 
```
