# Adding series to series

```python
import pandas as pd 
import numpy as np 
from pandas import Series,DataFrame
```

```python
ser_1 = Series(np.arange(4),index=list('abcd'))
ser_1
```

```python
ser_2 = Series(np.arange(5),index=list('adefg'))
ser_2
```

```python
ser_1 + ser_2  # index with not present in either of series addition on those indces will result in NaN
```

# Adding Two DataFrame with fill value

```python
df_1 = DataFrame(np.arange(16).reshape((4,4)),
                 index=list('abcd'),
                 columns=['col1','col2','col3','col4'])

df_1
```

```python
df_2 = DataFrame(np.arange(25).reshape(5,5),
                index=('a','b','c','d','e'),
                columns=['col1','col2','col3','col4','col5'])

df_2
```

```python
df_1 + df_2
```

```python
df_1.add(df_2,fill_value=0)
```

```python
df_2.add(df_1,fill_value=0)
```
