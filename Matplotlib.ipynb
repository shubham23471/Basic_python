```python
x = np.linspace(0,5,11)
y = x **2
```

### Matplotlib oops API 

```python
fig = plt.figure() # to make figure objects 

# this figure object acts as a blank canvas to plot we we have to axes first

axes = fig.add_axes([0.1,0.1,0.8,0.8])  # add axes nmber can be 0 to 1 
axes.plot(x,y)  # plot whatever you want
```

```python
fig1 = plt.figure()

axes1 = fig1.add_axes([0.1,0.1,0.89,0.9])
axes1.plot(x,y)
axes1.set_title('Outer Plot')

axes2 = fig1.add_axes([0.19,0.5,0.4,0.4])
axes2.set_title('Inner Plot')
axes2.plot(y,x,'r')
```

### Subplots (OOp)

```python
# doing the same thing as above using subplots 
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('X vs Y')

axes[1].plot(y,x)
axes[1].set_title('Y vs X')

plt.tight_layout()  # to avoid any overlappin between the plots
```

```python
# this is axes object refering to axes in just aabove cell.
# axes is an array of all the axes in the subplot and we can also iterate for this using for loop
# we can index the axes 
axes
```

```python
# understanding

fig100,axes100 = plt.subplots(nrows=1,ncols=2)

# we can iterate over the axes object

for num_of_axes in axes100:
    num_of_axes.plot(x,y)
    
# or we can index it 
  
```

### Figure size and DPI 

```python
fig22,axes22 = plt.subplots(nrows=2,ncols=1,figsize=(7,7))  # fig size in inches 

axes22[0].plot(x,y)

axes22[1].plot(y,x)
```

```python
fig23 = plt.figure()

ax = fig23.add_axes([0.8,0.1,0.6,0.7])
ax.plot(x,y,label='X vs Y')
ax.plot(y,x,label= 'Y vs X')

ax.legend(loc=(.5,.4))  # to label the lines 
                 # loc to move the label box use loc= 0 for let matplot lib the positon of label box
```

```python
# to change the figure size pass and argument figsize with tuple (row,column) length in the inches 

fig23 = plt.figure(figsize=(10,5))

ax = fig23.add_axes([0,0,1,1])
ax.plot(x,y,label='X vs Y')
ax.plot(y,x,label= 'Y vs X')

ax.legend()
```
